```python
import numpy as np
import pandas as pd

content = pd.read_csv("Content/content_cleaned.csv")
```

```python
from transformers import pipeline

fiction_categories = ["romance","science fiction","fantasy","horror","mystery","thriller","comedy","crime","historic", "tragedy", "politic", "drama", "arts"]

pipe = pipeline("zero-shot-classification", model="facebook/bart-large-mnli", device='cpu')
```

```python
content_with_caption = content[content["caption"].notna() & (content["caption"].str.strip() != "")]
sequence = content_with_caption["caption"].iloc[0]
```

```python
pipe(sequence, fiction_categories)
```

```python
max_index = np.argmax(pipe(sequence, fiction_categories)["scores"]) #On prend le score maximale parmis les probabilites possibles
max_label = pipe(sequence, fiction_categories)["labels"][max_index] #On trouve la categorie correspondant au score obtenu
max_label
```

```python
#On cree une fonction qui normalise le tout
def generate_predictions(sequence, categories):
    predictions = pipe(sequence, categories)
    max_index = np.argmax(predictions["scores"])
    max_label = predictions["labels"][max_index]
    return max_label
```

```python
content["categories"] = None
content
```

```python
id = []
predicted_cats = []

#On donne des categories au texte
missing_cats = content.loc[content["categories"].isna(), ["id", "caption"]].reset_index(drop=True)
```

```python
missing_cats
```

```python
from tqdm import tqdm

for i in tqdm(range(0, len(missing_cats))):
    sequence = missing_cats["caption"][i]
    predicted_cats += [generate_predictions(sequence, fiction_categories)]
    id += [missing_cats["id"][i]]
```

```python
missing_predicted_df = pd.DataFrame({"id": id, "predicted_categories": predicted_cats})
```

```python
missing_predicted_df
```

```python
content = pd.merge(content, missing_predicted_df, on="id", how="left")
content = content.drop(columns=["categories"])
```

```python
content.to_csv("Content/content_with_categories.csv", index=False)
```
