```python
import numpy as np
import pandas as pd

from datetime import datetime

content = pd.read_csv("Content/content.csv")
content = content.map(lambda x: str(x).replace('\r', '').strip())
```

```python
content
```

```python
#content["age_of_content"] = datetime.now() - pd.to_datetime((content["created_at"]))
```

```python
content_missing = content[~(content["caption"].isna()) |
                    ~(content["text"].isna())
]
# On enleve tous les contenus qui ne possede un "text" ou "caption" nul
content_missing
```

```python
content_missing["tagged_text"] = content_missing[["id", "text"]].astype(str).agg(": ".join,axis=1)
content_missing["tagged_caption"] = content_missing[["id", "caption"]].astype(str).agg(": ".join,axis=1)

#Pour facilement identifier le contenu
```

```python
content_missing
```

```python
{
    content_missing
    #.drop([], axis=1)
    .to_csv("Content/content_cleaned.csv", index=False)
}
#On enleve les elements superflues
#On cree content_cleaned.csv
```
